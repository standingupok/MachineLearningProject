```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_loss(h, y):
    epsilon = 1e-15
    h = np.clip(h, epsilon, 1 - epsilon)
    m = y.shape[0]
    return -1 / m * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))


def gradient_descent(X, y, theta, learning_rate, n_iterate):
    m = X.shape[0]
    losses = [] 
    for i in range(n_iterate):
        z = np.dot(X, theta)
        h = sigmoid(z)
        err = h - y
        gradient = (1/m) * np.dot(X.T, err)
        theta = theta - learning_rate * gradient

        loss = compute_loss(h, y)
        losses.append(loss)

        if(i % 100 == 0):
            print(f'Iteration {i} loss : {loss:.4f}')
    return theta, losses
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
X = np.array([[1, 2, 3],
              [1, 4, 9],
              [1, 6, 7]]) 
y = np.array([[0],
              [1],
              [1]])

learning_rate = 0.1
n_iterate = 1000
theta = np.zeros((X.shape[1],1))

theta_final, losses = gradient_descent(X, y, theta, learning_rate, n_iterate)
print("----------")
print(f'Final theta {theta_final}')
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ loss
plt.plot(range(len(losses)), losses, label="Loss")
plt.xlabel("Iterations")  # Nhãn trục x
plt.ylabel("Loss")        # Nhãn trục y
plt.title("Loss Over Iterations")  # Tiêu đề
plt.legend()
plt.grid()
plt.show()
```
